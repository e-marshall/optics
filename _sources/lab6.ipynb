6. net radiation equations

$$ E_{net} = E_{incoming SW} + E_{incoming LW} - M_{outgoing SW} - M_{outgoing LW}$$

We can disregard outgoing shortwave so net radiation becomes: 

$$ E_{net} = (1-a_{surface SW})E_{0}cos(\theta) +  (1-a_{surf LW}){\varepsilon}_{env} {\sigma}T_{env}^4 - {\varepsilon}_{surf}{\sigma}T_{surf}^4$$

                                                                      incoming SW           incoming LW        outgoing LW

for each jar: 

$$ E_{net glass} = (1-a_{surf SW})E_{0}cos(\theta) + (1-a_{surf LW}){\varepsilon}_{env}{\sigma}T_{env}^4 - {\varepsilon}_{glass}{\sigma}T_{glass}^4   $$

$$ E_{net foil} = (1-a_{surf SW})E_{0}cos(\theta) + (1-a_{surf LW}){\varepsilon}_{env}{\sigma}T_{env}^4 - {\varepsilon}_{foil}{\sigma}T_{glass}^4 $$

${\varepsilon}_{glass}$  is high,  ${\varepsilon}_{foil}$ is low, so ${\varepsilon}_{glass} > {\varepsilon}_{foil}$. The difference in emissivity will also impact incoming LW but the strongest effect is the impact of emissivity on outgoing LW, which we see reflected in the temp. v time curves.

7. Yes, using a thermal radiometer to measure the radiant temperatures of the objects at t=0 would show different temperatures between the two jars. The radiant temperature takes into account the  emitted radiation rather than the internal energy, so differences in emissivity will produce different radiant temperatures.
