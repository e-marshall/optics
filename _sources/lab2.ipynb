```python
import pandas as pd
import os
import holoviews as hv
import matplotlib.pyplot as plt
```

```python
spectra_df = pd.read_csv('221017_class_spectra.csv')
```

```python
spectra_df.head()
```

```python
spectra = spectra_df.drop('Wavelength', axis=1)
```

```python
spectra.plot(title = 'Spectral reflectance of all samples \n 10/17/2022 ~2:15 PM, Gardner Commons Quad',
            ylim=(0,5))
plt.legend(loc = 'center left', bbox_to_anchor=(1.0,0.5))
```

```python
sycamore = spectra_df[['221017.004','221017.005']]
```

```python
sycamore = sycamore.rename(columns = {'221017.004':'Sample 1',
                            '221017.005':'Sample 2'})
```

```python
sycamore
```

```python
sycamore.plot.line(title = 'Spectral reflectance of dry sycamore leaf \n 10/17/2022 Gardner Commons Quad, SLC, UT',
                   ylim=(0,2),xlim=(400,2200),ylabel = 'Reflectance', xlabel='Wavelength (nm)',alpha=0.5);
plt.axvspan(xmin=400, xmax=500, color='blue',alpha=0.2)
plt.axvspan(xmin=500, xmax=600, color='green', alpha=0.2)
plt.axvspan(xmin=600, xmax=700, color='red', alpha=0.2)
plt.axhline(y=1.0, color='black',alpha=0.2)
plt.axvline(x=1100,alpha=0.2)
plt.axvline(x=1650,alpha=0.2)
plt.axvline(x=1875,alpha=0.2)
```

The spectral reflectance profile shows very high reflectance across the spectrum. This is likely due to the concave shape of the leaf and the orientation of the sensor relative to the light source that caused a concentration of reflected radiation back toward the sensor. There appears to be noise at approximately 1000 nm and 2,150 nm each spanning ~ 50 nm. From ~ 1,500 - 1,600 nm there isa high degree of noise with reflectance ranging from about 0.5 to 1.75 in one sample and 0.75 to 5 in the second sample.

Outside of these features, there is no 'red-edge' indicating no photosynthetic activity which makes sense for a dead leaf. There appears to be a slight absorption feature centered around 1100 nm with a width of approximately 50 nm. There is a wider absorption feature between 1,650 nm and 1,875 nm. There also appears to be an absorption region from ~ 1,900 nm to 2,100 nm but the signal appears increasingly noisy moving into the higher wavelengths in this region. 

With the exception of the regions that are likely noise, reflection decreases as wavelength increases across the spectrum.
